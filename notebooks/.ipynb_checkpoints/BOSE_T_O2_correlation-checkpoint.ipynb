## Remove a space time mean

```python
T_200m = ds_T.THETA.sel(Z=-100, method='nearest')
```

```python
T_200m = T_200m.chunk({'YC':42, 'XC':40})
```

```python
T_200m.nbytes/1e9
```

```python
O2_200m = ds_O2.TRAC03.sel(Z=-50, method='nearest')
```

```python
O2_200m = O2_200m.chunk({'YC':42, 'XC':40})
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
T_200m.sel(XC=30, YC=-55, method='nearest').plot()

plt.subplot(122)
O2_200m.sel(XC=30, YC=-55, method='nearest').plot()
```

```python
# Detrend
T_200m_trend = T_200m.rolling(time=72, center=True).mean('time')
```

```python
T_200m_trend
```

```python
O2_200m_trend = O2_200m.rolling(time=72, center=True).mean()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
T_200m.sel(XC=30, YC=-55, method='nearest').plot()
T_200m_trend.sel(XC=30, YC=-55, method='nearest').plot()

plt.subplot(122)
O2_200m.sel(XC=30, YC=-55, method='nearest').plot()
O2_200m_trend.sel(XC=30, YC=-55, method='nearest').plot()
```

```python
T_detrend = T_200m - T_200m_trend
O2_detrend = O2_200m - O2_200m_trend
```

```python
plt.figure(figsize=(12,4))
plt.subplot(121)
T_detrend.sel(XC=30, YC=-55, method='nearest').plot()

plt.subplot(122)
O2_detrend.sel(XC=30, YC=-55, method='nearest').plot()
```

```python
# remove climatology as montly mean
gb = O2_detrend.groupby('time.month')
O2_clim = gb.mean(dim='time')
O2_clim_anom = gb - O2_clim
```

```python
gb = T_detrend.groupby('time.month')
T_clim = gb.mean(dim='time')
T_clim_anom = gb - T_clim
```

```python
O2_clim_anom
```

```python
O2_clim.sel(XC=30, YC=-55, method='nearest').plot()
```

```python
T_clim.sel(XC=0, YC=-40, method='nearest').plot()
T_clim.sel(XC=0, YC=-50, method='nearest').plot()
```

```python
T_clim.isel(month=0).plot(vmin=-1)
```

```python
gb = T_detrend.rolling(XC=30, YC=30, center=True).mean().groupby('time.month')
T_clim_sm = gb.mean(dim='time')
T_clim_sm_anom = gb - T_clim_sm
```

```python
gb = O2_detrend.rolling(XC=30, YC=30, center=True).mean().groupby('time.month')
O2_clim_sm = gb.mean(dim='time')
O2_clim_sm_anom = gb - O2_clim_sm
```

```python
T_clim_sm.isel(month=0).plot(vmin=-1)
```

```python
T_clim_sm_anom
```

```python
O2_clim_sm_anom
```

```python
T_clim_sm_anom.sel(XC=30, YC=-40, method='nearest').plot()
```

```python
O2_clim_sm_anom.sel(XC=30, YC=-40, method='nearest').plot()
```

```python
Tvals = T_clim_sm_anom.sel(XC=30, YC=-40, method='nearest').values
O2vals = O2_clim_sm_anom.sel(XC=30, YC=-40, method='nearest').values
```

```python
plt.scatter(Tvals, 
            O2vals , s =1)
```

```python
T_clim_sm_anom
```

```python
T_var = (T_clim_sm_anom**2).mean('time')
O2_var = (O2_clim_sm_anom**2).mean('time')
```

```python
T_var
```

```python
TO2_covar = (T_clim_sm_anom*O2_clim_sm_anom).mean('time')
```

```python
TO2_covar
```

```python
corr_coeff = TO2_covar / T_var**0.5 / O2_var**0.5
```

```python
corr_coeff.load()
```

```python
corr_coeff.plot()
```

```python
corr_coeff.plot.hist(bins=np.linspace(-1,1,50));
plt.title('Corr coeff T and O2 100m')
```
